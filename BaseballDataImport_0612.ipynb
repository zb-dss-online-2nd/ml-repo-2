### data import_월별

```python
datas=[]

for year in years:
    driver.find_element_by_css_selector('#ddlYear').send_keys(year)
    WebDriverWait(driver, 20).until(EC.presence_of_all_elements_located)
    print('연도', year)
    for month in months:
        driver.find_element_by_css_selector('#ddlMonth').send_keys(month)
        WebDriverWait(driver, 20).until(EC.presence_of_all_elements_located)
        option = driver.find_element_by_xpath("//*[text()='%s']"%month)
        driver.execute_script("arguments[0].scrollIntoView();", option)
        option.click()
        print('월별',month)

        page=driver.page_source
        soup=BeautifulSoup(page, "html.parser")
        table_text=soup.select_one('#contents > table').text
        table_text=table_text.lstrip('\n\n\n\n\n\n\n\n\n\n\n\n\n일\n월\n화\n수\n목\n금\n토\n\n\n1')
        print(table_text)

        #이상으로 table_text를 추출
        
        pattern = re.compile(r'[^0-9|\]|\s]\w+\s\d+\s:\s\d+\s\w+')
        scores = pattern.findall(table_text)
        scores= list( map(lambda x: x.strip(), scores) )

        #월별 구단종류
        team_set={}
        team_set=set()
        team_sets={}

        #월별 수치 초기화
        team_wins={}
        all_wins=0
        win_rate={} 
        team_games={}

        # 팀목록

        for score in scores:
            print(score)
            teamA, teamB = score.split(':')
            teamA= teamA.split()[0]
            teamB= teamB.split()[1]
            print(teamA, teamB)
            team_set.add(teamA)
            team_set.add(teamB)
        print('이번달 참전 팀:', team_set)

        
        
        for team in team_set:
            team_wins[team] = 0
            team_games[team] = 0
        
        all_wins=0


        for score in scores:
            print( score )
            scoreA, scoreB = score.split(':')
            scoreA_point= scoreA.split()[1]
            scoreB_point= scoreB.split()[0]
            teamA=scoreA.split()[0]
            teamB=scoreB.split()[1]
            print('left team=', teamA, '점수=', scoreA_point,'right team', teamB, '점수=', scoreB_point)   
            if scoreA_point > scoreB_point:
                team_wins[teamA] +=1
                team_games[teamA] +=1
                team_games[teamB] +=1
                all_wins+=1
            elif scoreA_point < scoreB_point:
                team_wins[teamB] +=1
                team_games[teamA] +=1
                team_games[teamB] +=1
                all_wins+=1
        
        

        for team in team_set:
            win_rate[team] = round( team_wins[team]/team_games[team],3 ) 
        
        print(team_wins)
        print(win_rate)
           
        
        for team in win_rate:
            datas.append(

                {'year': year, 'month' : month, 'win_rate[%s]'%team :win_rate[team]}      )
```
